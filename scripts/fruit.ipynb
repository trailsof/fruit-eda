```python
fresh_fruit["CaloriesPerPound"] = fresh_fruit["Calories_per_100g"] * 4.536
fresh_fruit["CaloriesPerDollar"] = fresh_fruit["CaloriesPerPound"] / fresh_fruit["RetailPrice"]
fresh_fruit[["CaloriesPerDollar", "Fruit"]].head(2)
```

```python
fresh_fruit_sorted = fresh_fruit.sort_values("CaloriesPerDollar", ascending=False)
```

```python
fresh_fruit_sorted.plot(
    kind="barh", 
    x="Fruit", 
    y="CaloriesPerDollar", 
    legend=False, 
    figsize=(8,6),
    title="Calories per Dollar Fresh Fruit"
)
plt.xlabel("Calories per Dollar")
plt.show()
```

```python

yield_sorted = fresh_fruit.sort_values("Yield", ascending=False)

# Define colors: green ≥0.7, yellow 0.5–0.7, red <0.5
colors = [
    "seagreen" if y >= 0.7 else ("gold" if y >= 0.5 else "tomato")
    for y in yield_sorted["Yield"]
]

plt.figure(figsize=(12,6))
plt.barh(yield_sorted["Fruit"], yield_sorted["Yield"], color=colors)

plt.xlabel("Yield (edible pounds per pound purchased)")
plt.title("Fresh fruit yield (highlighting high- and medium-waste fruits)")
plt.gca().invert_yaxis()  # highest yield at top

# Add value labels
for i, v in enumerate(yield_sorted["Yield"]):
    plt.text(v + 0.01, i, f"{v:.2f}", va="center")

# Custom legend
legend_elements = [
    Patch(facecolor="seagreen", label="High yield (≥ 0.7)"),
    Patch(facecolor="gold", label="Medium yield (0.5–0.7)"),
    Patch(facecolor="tomato", label="Low yield (< 0.5)")
]
plt.legend(handles=legend_elements, loc="lower right")

plt.show()
```

```python
fresh_fruit["PricePerEdibleLb"] = fresh_fruit["RetailPrice"] / fresh_fruit["Yield"]
```

```python
# Cups per lb (edible)
fresh_fruit.loc[:, "CupsPerLb"] = fresh_fruit["Yield"] / fresh_fruit["CupEquivalentSize"]

# Recalculate cup price
fresh_fruit.loc[:, "CalcCupPrice"] = fresh_fruit["PricePerEdibleLb"] / fresh_fruit["CupsPerLb"]

# Cups per dollar (factoring in yield)
fresh_fruit.loc[:, "CupsPerDollar"] = 1 / fresh_fruit["CalcCupPrice"]
```

```python
sorted_cups = fresh_fruit.sort_values("CupsPerDollar", ascending=False)

plt.figure(figsize=(12,6))
plt.barh(sorted_cups["Fruit"], sorted_cups["CupsPerDollar"], color="orchid")

plt.xlabel("Cups per $1 (edible, yield-adjusted)")
plt.title("How many cups of fruit you can buy for $1 (factoring in waste)")
plt.gca().invert_yaxis()

# Add labels
for i, v in enumerate(sorted_cups["CupsPerDollar"]):
    plt.text(v + 0.05, i, f"{v:.2f} cups", va="center")

plt.show()
```

```python
categories = {
    "Berry": ["Strawberries", "Blueberries", "Blackberries", "Raspberries"],
    "Citrus": ["Oranges", "Grapefruit"],
    "Tropical": ["Bananas", "Mangoes", "Pineapple", "Papaya", "Kiwi"],
    "Stone": ["Peaches", "Plum", "Cherries"],
    "Pome": ["Apples", "Pears"],
    "Melon": ["Watermelon", "Cantaloupe", "Honeydew"],
    "Other": ["Pomegranate"]
}

category_map = {fruit: cat for cat, fruits in categories.items() for fruit in fruits}
fresh_fruit["Category"] = fresh_fruit["Fruit"].map(category_map).fillna("Other")
```

```python
fresh_fruit["Category"]
```

```python
# Group by category and calculate mean CupsPerDollar
category_summary = fresh_fruit.groupby("Category").agg({
    "CupsPerDollar": "mean",
    "Fruit": "count"
}).rename(columns={"Fruit": "NumFruits"}).reset_index()
category_summary
```

```python
# ---- use your existing summary ----
cs = category_summary.sort_values("CupsPerDollar", ascending=True).reset_index(drop=True)
y = np.arange(len(cs))
x = cs["CupsPerDollar"].to_numpy()
labels = [f"{c}  (n={n})" for c, n in zip(cs["Category"], cs["NumFruits"])]

# ---- layout: bars on left, legend on right ----
fig = plt.figure(figsize=(12, 6))
gs = fig.add_gridspec(1, 2, width_ratios=[3, 1], wspace=0.3)
ax  = fig.add_subplot(gs[0, 0])
axr = fig.add_subplot(gs[0, 1]); axr.axis("off")

# ---- colors mapped to value (viridis) ----
cmap = plt.get_cmap("viridis")
norm = mpl.colors.Normalize(vmin=x.min(), vmax=x.max())
colors = cmap(norm(x))

# ---- bars ----
bars = ax.barh(y, x, color=colors)
ax.set_yticks(y, labels)
ax.set_xlabel("Mean cups per $")
ax.set_title("Cups per dollar by category")
ax.grid(axis="x", linestyle="--", alpha=0.3)
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

# value labels: inside for long bars, outside for short ones
inside_cut = x.max() * 0.18
for yi, xi in enumerate(x):
    if xi >= inside_cut:
        ax.text(xi - x.max()*0.02, yi, f"{xi:.2f}", ha="right", va="center", color="white", fontweight="bold")
    else:
        ax.text(xi + x.max()*0.02, yi, f"{xi:.2f}", ha="left", va="center")

ax.set_xlim(0, max(0.01, x.max()*1.15))

# ---- compact legend of fruit types per category ----
legend_lines = []
for cat in cs["Category"]:
    fruits = sorted(set(fresh_fruit.loc[fresh_fruit["Category"] == cat, "Fruit"].astype(str)))
    legend_lines.append(cat + ":\n  " + textwrap.fill(", ".join(fruits), width=28))
axr.text(0, 1, "\n\n".join(legend_lines), ha="left", va="top", fontsize=9)

# optional colorbar to explain colors
sm = mpl.cm.ScalarMappable(cmap=cmap, norm=norm); sm.set_array([])
cbar = fig.colorbar(sm, ax=axr, fraction=0.06, pad=0.08)
cbar.set_label("Mean cups per $")

plt.tight_layout()
plt.show()
```
